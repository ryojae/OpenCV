```
import cv2
import numpy as np

# 이미지 읽기
img_path = "rgb.png"
image = cv2.imread(img_path)

plt.imshow(image)
```

```
# BGR을 HSV 컬러 스페이스로 변환
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# 흰색 영역 검출
lower_color = np.array([0, 0, 200], dtype=np.uint8)
upper_color = np.array([255, 30, 255], dtype=np.uint8)


color_mask = cv2.inRange(hsv, lower_color, upper_color)       # 범위내의 픽셀들은 흰색, 나머지 검은색
color_result = cv2.bitwise_and(image, image, mask=color_mask) # 바이너리 이미지를 마스크로 사용하여 원본이미지에서 범위값에 해당하는 영상부분을 획득
plt.imshow(color_result)
```
